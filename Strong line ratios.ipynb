```python
def rfactor(linedata, contdata, A=8.659):
    return 1.0 + 1.0/(A*linedata/contdata - 1.0)
```

```python
full487 = fits.open("../F487N.fits")[0]
full656 = fits.open("../F656N.fits")[0]
full658 = fits.open("../F658N.fits")[0]
full502 = fits.open("../F502N.fits")[0]
full547 = fits.open("../F547M-cr.fits")[0]

r487 = rfactor(full487.data, full547.data, 8.659)
r656 = rfactor(full656.data, full547.data, 43.16)
r658 = rfactor(full658.data, full547.data, 26.06)
r502 = rfactor(full502.data, full547.data, 8.592)

fits.PrimaryHDU(header=full487.header, data=r487).writeto("rcorrect-full-F487N.fits")
fits.PrimaryHDU(header=full656.header, data=r656).writeto("rcorrect-full-F656N.fits")
fits.PrimaryHDU(header=full658.header, data=r658).writeto("rcorrect-full-F658N.fits")
fits.PrimaryHDU(header=full502.header, data=r502).writeto("rcorrect-full-F502N.fits")


```

```python
def mixline(N1, O1, N2, O2, n=100):
    """
    Calculate the BPT ratio for a mixture of two pure states
    """
    frac = np.linspace(0.0, 1.0, n)
    rN = (10**N1) * frac + (10**N2) * (1. - frac)
    rO = (10**O1) * frac + (10**O2) * (1. - frac)
    return np.log10(rN), np.log10(rO)
```

```python
def ifracline(U, n=50):
    """
    BPT ratio for a sequence of ionization parameter: U
    
    Assume that n(++) / n(+) = U
    """
    nsingle = 1.0/(1.0 + U)
    ndouble = U/(1.0 + U)
    return np.log10(nsingle), np.log10(ndouble)
    
```

### T-sensitivity of strong line ratios from pyneb

```python
import pyneb
hi = pyneb.RecAtom("H", 1)
nii = pyneb.Atom("N", 2)
oiii = pyneb.Atom("O", 3)
```

```python
def nii_ha(T, den, abun=7.73):
    A = nii.getEmissivity(T, den, wave=6584)
    B = hi.getEmissivity(T, den, label="3_2")
    return np.log10(A/B) + abun - 12.0

def oiii_hb(T, den, abun=8.67):
    A = oiii.getEmissivity(T, den, wave=5007)
    B = hi.getEmissivity(T, den, label="4_2")
    return np.log10(A/B) + abun - 12.0

def ha_hb(T, den):
    A = hi.getEmissivity(T, den, label="3_2")
    B = hi.getEmissivity(T, den, label="4_2")
    return A/B

```

```python
Trange = np.linspace(7000.0, 15000.0)
denrange = [1000., 1e4, 1e5]
colors = "gbr"
for den, c in zip(denrange, colors):
    plot(Trange, 1.0 + nii_ha(Trange, den), c + '-', 
         label='10 x 6583/6563, den = {:.0e}'.format(den))
    plot(Trange, oiii_hb(Trange, den), c + '--', 
         label='5007/4861, den = {:.0e}'.format(den))
grid()
legend(ncol=3)
xlabel("Temperature, K")
ylabel("log 10(Line ratio)")
gcf().set_size_inches(12, 8)
```

```python
def bpt_plot(deltaT=0.0, Oabun=8.5, Nabun=7.65, den=5000.0):

    plt.imshow(H.T**.5, extent=[xmin, xmax, ymin, ymax], 
               origin='lower', interpolation='none', aspect='equal',
               cmap=cm.gray_r
               )
    grid()
    xx = np.linspace(xmin, xmax, 100)
    for y0 in -0.1, 0.0, 0.1, 0.2:
        plot(xx, y0 - 0.5*xx, 'r:', alpha=0.5) 
    for y0 in 1.0 + 0.2*np.arange(8):
        plot(xx, y0 + 2.0*xx, 'r:', alpha=0.5) 

    Urange = np.logspace(-1.0, 1.0, 200)
    x, y = ifracline(Urange)
    colors = "bgrc"
    for T, c in zip([7500, 8000, 8500, 9000], colors):
        plot(x + nii_ha(T+deltaT, den, abun=Nabun), y + oiii_hb(T, den, abun=Oabun), 
             c + '-', alpha=0.8,
             label="T(++) = {:.0f} K".format(T))
    Trange = np.linspace(5000, 12000, 200)
    for U, c in zip([0.5, 1.0, 2.0, 5.0], colors):
        x, y = ifracline(U)
        plot(x + nii_ha(Trange+deltaT, den, abun=Nabun), y + oiii_hb(Trange, den, abun=Oabun), 
             c + '--', alpha=0.8,
             label="U = {:.1f}".format(U))
    xlabel("log10 ([N II] 6583 / Ha 6563)")
    ylabel("log10 ([O III] 5007 / Hb 4861)")
    xlim(xmin, xmax)
    ylim(ymin, ymax)
    legend(ncol=2)
    title("n = {:.1e} pcc, Abundances = O:{:.2f}, N:{:.2f}, T(+) = T(++) + {:.0f} K".format(den, Oabun, Nabun, deltaT))
    gcf().set_size_inches(7, 7)

bpt_plot()
gcf().savefig('BPT-standard.pdf')
```
