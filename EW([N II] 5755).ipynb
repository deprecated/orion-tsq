### Function to find equivalent widths

```python
def find_ew(wavs, spec, wav1, wav2, av_cont=None):
    """Calculate equivalent width of a line
    wav1 -> wav2 is assumed to be the line, while wav < wav1 and wav > wav2
    is assumed to be the continuum, unless av_cont is set"""
    if av_cont is None:
        av_cont = spec[(wavs < wav1) | (wavs > wav2)].mean()
    excess = (spec/av_cont) - 1.0
    return np.trapz(excess, wavs)
```

### EW in Ring Nebula from SPM spectra

Bottom line is EW([N II] 5755) ~= 100 Å, whereas nominal width of FQ575N is 20.6 Å, so continuum contamination is only 20%.

```python
rhdu, = fits.open("/Users/will/Work/RingNebula/WFC3/2013-Geometry/Spectra/E1200PA150-bg02.fits")
```

```python
nx, wav0, i0, dwav = [rhdu.header[k] for k in "NAXIS1", "CRVAL1", "CRPIX1", "CD1_1"]
wavs = wav0 + (np.arange(nx) - (i0 - 1))*dwav 
```

Wide window `jw1`–`jw2` to include some continuum.  Narrow window `j1`–`j2` is mainly line.

```python
jw1 = wavs[wavs < 5720.0].argmax()
jw2 = wavs[wavs < 5790.0].argmax()
j1 = wavs[wavs < 5740.0].argmax()
j2 = wavs[wavs < 5770.0].argmax()
jw1, j1, j2, jw2
```

```python
plot(4*rhdu.data[:,jw1:j1].mean(axis=-1), label="4 x cont blue")
plot(rhdu.data[:,j1:j2].mean(axis=-1), label="[N II] 5755")
plot(4*rhdu.data[:,j2:jw2].mean(axis=-1), label="4 x cont red")
legend()
xlim(100., 220.)
ylim(-0.1, 1.0)
gcf().set_size_inches(15,6)
```

```python
spec_ringA = rhdu.data[132:138,:].sum(axis=0)
spec_ringB = rhdu.data[184:190,:].sum(axis=0)
spec_innerA = rhdu.data[140:150,:].sum(axis=0)
spec_innerB = rhdu.data[174:184,:].sum(axis=0)
plot(wavs, spec_ringA + spec_ringB, label="outer")
plot(wavs, spec_innerA + spec_innerB, label="inner")
ylim(0.0, 20.)
xlim(5720, 5800)
legend()
gcf().set_size_inches(15,6)
```

```python
window = (wavs > 5720.0) & (wavs < 5790.0)
ew_outer = find_ew(wavs[window], spec_ringA[window] + spec_ringB[window], 5740.0, 5770.0)
ew_inner = find_ew(wavs[window], spec_innerA[window] + spec_innerB[window], 5740.0, 5770.0)
ew_outer, ew_inner
```

So EW([N II]) is rather large, especially in the outer ring.

Now check that the H beta EW comes out OK.

```python
plot(wavs, spec_ringA + spec_ringB, label="outer")
plot(wavs, spec_innerA + spec_innerB, label="inner")
ylim(0.0, 10.)
xlim(4500, 4800)
gcf().set_size_inches(15,6)
```

Finding some clean continuum is a bit tricky, but we do manage it.

```python
window = (wavs > 4830.0) & (wavs < 4890.0)
cont_window = (wavs > 4590.0) & (wavs < 4610.0)
ew_outer = find_ew(wavs[window], spec_ringA[window] + spec_ringB[window], 
                   4845.0, 4875.0, (spec_ringA[cont_window] + spec_ringB[cont_window]).mean())
ew_inner = find_ew(wavs[window], spec_innerA[window] + spec_innerB[window], 
                   4845.0, 4875.0, (spec_innerA[cont_window] + spec_innerB[cont_window]).mean())
ew_outer, ew_inner
```

So these are reasonable.
