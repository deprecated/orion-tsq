## Extinction with foreground emission

Fraction $x$ of the emission comes from in front of a screen of depth $\tau$ at H$\beta$.  The extinction ratio from H$\alpha$ to H$\beta$ is $(1 + f_{\lambda})$. 

```python
def A(t, x):
    return (1.-x)*np.exp(-t) + x

def B(t, x, flam):
    return A(t*(1.+flam), x) / A(t, x)

def tau_extinct(extinct):
    return -np.log(extinct)

def tau_redden(decrement, flam):
    return -np.log(decrement)/flam

colors = "bgrc"
tau = np.linspace(0.0, 10.0)
flam = -0.22
for x, c in zip([0.01, 0.05, 0.1, 0.2], colors):
    plot(tau, tau_redden(B(tau, x, flam), flam), c, label="frac = {:.2f}".format(x))
    plot(tau, tau_extinct(A(tau, x)), c + '--')
legend(loc="upper left")
ylim(0.0, 5.0)
xlim(0.0, 10)
xticks(range(11))
xlabel("True optical depth, tau")
ylabel("Apparent optical depth")
grid()
gcf().set_size_inches(10, 5)
```

The solid line shows the apparent optical depth calculated from Balmer decrement reddening.  The dashed line shows the apparent optical depth calculated from extinction (either Ha-to-radio, or from spatial brightness profile).
