So the above is sort of promising.  A linear fit looks kind of reasonable and has a determination coefficient of 20%.  Some of the remaining variance is probably due to noise, and then what is left in the middle might be the $t^2$ variations that we are looking for...  The linear fit is tugged upwards a bit by the skew in the  575/658 distribution.

We can work out this a bit more carefully. 

````python
W547, W658, W575 = 169.0, 7.3, 4.7
Tn, Ta = 0.26, 0.23
k658 = 0.8  # We don't really know this one yet

def EWa(R575, k575=0.938):
    return (W547/(k575*R575) - W575)/Ta

def EWn(R658):
    return (W547/(k658*R658) - W658)/Tn

def ratio_nii(R, R575, R658, k575=0.938):
    """[N II] 5755/6584 ratio"""
    return R * (Tn/Ta) * (1.0 - W575*k575*R575/W547) / (1.0 - W658*k658*R658/W547)

#def ratio_nii(R, R575, R658):
#    return R * (Tn/Ta) * (1.0 + W658/(Tn*EWn(R658))) / (1.0 + W575/(Ta*EWa(R575)))
````

````python
def ratio_vs_ewneb(k575=0.938):
    ew5755 = EWa(r547_575, k575=k575)
    ew6584 = EWn(r547_658)
    rnii = ratio_nii(r575_658, r547_575, r547_658, k575=k575)
    x = ew6584[m]
    y = rnii[m]
    xm, ym = np.median(x), np.median(y)
    x1, x2 = 0.5*xm, 3.5*xm
    y1, y2 = -0.1*ym, 2.0*ym
    mm = (x > x1) & (x < x2) & (y > y1) & (y < y2) 
    plot(x, y, "k.", ms=0.03, zorder=100)
    xlim(x1, x2)
    ylim(y1, y2)
    xlabel("EW([N II] 6584), Angstrom")
    ylabel("[N II] 5755 / 6584")
    print "{} = {:.4f} +/- {:.4f}, {} = {:.4f} +/- {:.4f}".format(
                        "EW", np.median(x[mm]), np.std(x[mm]), 
                        "5755/6584", np.median(y[mm]), np.std(y[mm]))
                        
    slope, intercept, r_value, p_value, std_err = scipy.stats.linregress(x[mm], y[mm])
    print "Linear fit: y = {:.3f} (x/1000) + {:.3f}, r = {:.3f}, p = {:.6f}, R^2 = {:.3f}".format(
                    1000*slope, intercept, r_value, p_value, r_value**2)
    xx = np.linspace(x1, x2)
    plot(xx, slope*xx + intercept, "r--", lw=3, alpha=0.5)
    grid()
    gcf().set_size_inches(18,9)
````

````python
def plot_hist_nii_ratio_vs_ewneb(k575=0.938, 
                                 ewranges=[[0, 500], [500, 750], [750, 1000], [1000, 1400], [1400, 3000]]):
    ew5755 = EWa(r547_575, k575=k575)
    ew6584 = EWn(r547_658)
    rnii = ratio_nii(r575_658, r547_575, r547_658, k575=k575)
    delta = 0.02
    x = ew6584[m]
    y = rnii[m]
    weights = s658[m]
    y1, y2 = -0.005, 0.05
    offset = 0.0
    yp = ratio_nii(r575_658, r547_575, r547_658, k575=k575 + delta)[m]
    yn = ratio_nii(r575_658, r547_575, r547_658, k575=k575 - delta)[m]
    for x1, x2 in ewranges:
        mm = (x > x1) & (x < x2) & (y > y1) & (y < y2)
        hist, edges = np.histogram(y[mm], bins=200, weights=weights[mm], 
                                   range=[y1, y2], density=True)
        norm = hist.max()
        hist /= norm
        histp, _ = np.histogram(yp[mm], bins=200, weights=weights[mm], 
                                range=[y1, y2], density=True)
        histn, _ = np.histogram(yn[mm], bins=200, weights=weights[mm],
                                range=[y1, y2], density=True)
        histp /= norm
        histn /= norm
        centers = 0.5*(edges[:-1] + edges[1:])
        label = "EW(6584) = {}-{} A".format(x1, x2)
        plot(centers, hist + offset, label=label, drawstyle="steps-mid", alpha=1.0)
        fill_between(centers, histn + offset, histp + offset, alpha=0.15, facecolor="k")
        offset -= 0.6
    legend()
    grid()
    xlabel("[N II] 5755/6584")
    ylabel("Relative distribution")
    ylim(-2.5, 1.05)
    gcf().set_size_inches(8, 8)

plot_hist_nii_ratio_vs_ewneb()
gcf().savefig("histogram_nii_ratio_vs_ewneb.pdf")
````

````python
def Tdiag_plot(atom, wavA, wavB, temrange, denrange):
    for den in denrange:
        A = atom.getEmissivity(temrange, den, wave=wavA)
        B = atom.getEmissivity(temrange, den, wave=wavB)
        plot(temrange, A/B, label="n = {:.0e} pcc".format(den))
    legend(loc="lower right")
    xlabel("Temperature, K")
    ylabel("{} / {}".format(wavA, wavB))

Tdiag_plot(nii, 5755, 6583, np.linspace(5000.0, 1.8e4), 
           [500.0, 1.e3, 2.e3, 4.e3, 8.e3])
ylim(0.0, 0.04)
grid()
````

### Logarithmic sensitivity of R([N II]) to T and n

We need to find $d \log R / d \log T$ so we can convert ${\sigma_R}^2 /R_0^2 $ into ${\sigma_T}^2 / T_0^2 \equiv t^2$. 

````python
def rnii_func(T, den):
    A = nii.getEmissivity(T, den, wave=5755)
    B = nii.getEmissivity(T, den, wave=6583)
    return A/B

T0s = [1.e4, 1.2e4]
den0s = [2.e3, 5.e3]
for T0, den0 in zip(T0s, den0s):
    t = np.logspace(-1.0, 1.0)
    r = rnii_func(t*T0, den0) / rnii_func(T0, den0)
    loglog(t, r, label="T0 = {:.0f} K, n = {:.0f} pcc".format(T0, den0))

for a in [1.0, 1.5, 2.0]:
    loglog(t, t**a, '--', label="(T/T0)^{:.1f}".format(a))

xlim(0.1, 10.0)
ylim(0.1, 10.0)
xlabel("T/T0")
ylabel("R/R0")
legend(loc="upper left")
grid()
gcf().set_size_inches(6, 6)
````

````python
#dens = [1000, 2000, 5000, 8000]
ax1 = gca()
dens = [1000, 1e4, 1e5, 1e6]
Trange = np.linspace(4000.0, 20000.0, 200)
for den in dens:
    Rs = rnii_func(Trange, den)
    s = np.diff( np.log(Rs) ) / np.diff( np.log(Trange) )
    ax1.plot(Rs[:-1], s, label="n = {:.0e} pcc".format(den))

ax2 = ax1.twinx()    
Ts = [8000, 1e4, 1.2e4, 1.4e4]
denrange = np.logspace(2.0, 7.0, 200)
for T in Ts:
    Rs = rnii_func(T, denrange)
    s = np.diff( np.log(Rs) ) / np.diff( np.log(denrange) )
    ax2.plot(Rs[:-1], s, '--', label="T = {:.0f} K".format(T))

    
ax1.legend(loc="upper center", title="d ln R / d ln T")
ax1.grid()
ax1.set_xlim(0.0, 0.05)
ax1.set_ylim(0.0, 5.0)
ax1.set_xlabel("R([N II]) = 5755 / 6583")
ax2.set_ylabel("Sensitivity to temperature: s(T) = d ln R([N II]) / d ln T")
    
ax2.set_ylim(0.0, 5.0)
ax2.legend(loc="center left", title="d ln R / d ln n", bbox_to_anchor=(0, 0, 1.0, 0.5))
ax1.set_ylabel("Sensitivity to density: s(n) = d ln R([N II]) / d ln n")

gcf().set_size_inches(7, 7)
gcf().savefig("rnii-sensitivity.pdf")
````

````python
def two_phase_rsii_rnii(den_hi, f_hi, den_lo=1000.0, T0=8000.0, T_hi=None):
    """Calculate [S II] and [N II] ratios from mixture of 2 densities"""
    if T_hi is None:
        T_hi = T0
    As_hi = sii.getEmissivity(T_hi, den_hi, wave=6716) 
    As_lo = sii.getEmissivity(T0, den_lo, wave=6716) 
    Bs_hi = sii.getEmissivity(T_hi, den_hi, wave=6731)
    Bs_lo = sii.getEmissivity(T0, den_lo, wave=6731)
    rsii = (f_hi*As_hi + (1.-f_hi)*As_lo)/(f_hi*Bs_hi + (1.-f_hi)*Bs_lo)
    
    An_hi = nii.getEmissivity(T_hi, den_hi, wave=5755) 
    An_lo = nii.getEmissivity(T0, den_lo, wave=5755) 
    Bn_hi = nii.getEmissivity(T_hi, den_hi, wave=6583)
    Bn_lo = nii.getEmissivity(T0, den_lo, wave=6583)
    rnii = (f_hi*An_hi + (1.-f_hi)*An_lo)/(f_hi*Bn_hi + (1.-f_hi)*Bn_lo)
    return rsii, rnii

two_phase_rsii_rnii(1.e6, np.linspace(0.0, 1.0, 11))
    
    
````
