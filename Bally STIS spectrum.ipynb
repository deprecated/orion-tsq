## Use the Bally STIS spectrum of HST10

```python
from astropy.io import fits
```

```python
hdulist = fits.open("bally-stis-hst10-g430l-cr.fits")
hdulist.info()
```

```python
hdu = hdulist["SCI"]
nx, wav0, i0, dwav = [hdu.header[k] for k in "NAXIS1", "CRVAL1", "CRPIX1", "CD1_1"]
wavs = wav0 + (np.arange(nx) - (i0 - 1))*dwav 
ny, x0, j0, dx = [hdu.header[k] for k in "NAXIS2", "CRVAL2", "CRPIX2", "CD2_2"]
xpos = x0 + (np.arange(ny) - (j0 - 1))*dx 
```

```python
import aplpy
```

```python
f = aplpy.FITSFigure(hdu)
f.show_grayscale(vmin=3e-16, stretch="log")
f.show_colorbar()
```

### Set up the filters

```python
filters = {}
filters["FQ575N"] = {"wav0": 5758.0, "dwav": 20.6}
filters["F547M"] = {"wav0": 5547.0, "dwav": 650.0}
filters["F547M red"] = {"wav0": 5547.0 + (650./4.), "dwav": 650.0/2.}
filters["F487N"] = {"wav0": 4871.0, "dwav": 62.0}
filters["F502N"] = {"wav0": 5010.0, "dwav": 65.0}
filters["F469N"] = {"wav0": 4688.0, "dwav": 53.1}
filters["FQ437N"] = {"wav0": 4371.0, "dwav": 33.0}
filters
```

```python
for name, params in filters.items():
    if "wav0" in params:
        params["i1"] = len(wavs[wavs < params["wav0"] - 0.5*params["dwav"]])
        params["i2"] = len(wavs[wavs <= params["wav0"] + 0.5*params["dwav"]]) + 1
filters
```

```python
def mean_profile(image, filt, red_edge=1110, blue_edge=93):
    i1 = max(filt["i1"], blue_edge)
    i2 = min(filt["i2"], red_edge)
    return image[:, i1:i2].mean(axis=1)
```

The FQ575N filter doesn't quite fit in the G430L spectrum, so do another one next to it. 

```python
# filters["FQ575N shift"] = {'i1': 1075, 'i2': 1108} # Broad
filters["FQ575N shift"] = {'i1': 1090, 'i2': 1108} # Narrow
#filters["FQ575N shift"] = {'i1': 1100, 'i2': 1108} # Very Narrow
filters["FQ575N double shift"] = {'i1': 1070, 'i2': 1090} # Just to blue
```

Also the F469N filter can not be well simulated, since the blended He I 4713 and [Fe III] 4658 lines get in the way.  These are both on the edge of the filter bandpass, so it is not clear how much they enter (we suspect that the [Fe III] line does enter though).  In the spectrum, however, due to the 2 arcsec (= 100 Angstrom) slit width, the lines contribute about 5-10% of the filter flux (mainly [Fe III]).    

So, we look slightly to the blue, where it looks like pure continuum. 

```python
filters["F469N shift"] = {'i1': 690, 'i2': 708}
```

### Look at the 1D spectrum

```python
def filter_box(filt, **kwds):
    x1 = wavs[filt["i1"]]
    x2 = wavs[filt["i2"]]
    fill_betweenx([0.0, 1.0e-15], [x1, x1], [x2, x2], **kwds)
```

```python
plot(wavs, hdu.data[93:360, :].mean(axis=0))
plot(wavs, hdu.data[378:579, :].mean(axis=0))
plot(wavs, hdu.data[617:823, :].mean(axis=0))
plot(wavs, hdu.data[845:1089, :].mean(axis=0))
filter_box(filters["F469N"], alpha=0.1)
filter_box(filters["F469N shift"], alpha=0.1)
filter_box(filters["F547M"], alpha=0.1)
filter_box(filters["FQ575N shift"], alpha=0.1)
filter_box(filters["FQ575N double shift"], alpha=0.1)

ylim(0.0, 1e-15)
xlim(4500, 5800)
grid()
xlabel("Wavelength, Angstrom")
gcf().set_size_inches(20,8)
```

### Compare profiles in the filters

This is the full continuum filter

```python
f547m = mean_profile(hdu.data, filters["F547M"])
```

And this is the red half, that we can see in the Rubin spectra.

```python
f547m_rubin = mean_profile(hdu.data, filters["F547M red"])
```

And this is the narrow filter close to the red edge, which is as close as we can get to the 5755 line

```python
fq575n = mean_profile(hdu.data, filters["FQ575N shift"])
```

Now the F469N filter - the real wave range and the closest patch of pure continuum.

```python
f469n = mean_profile(hdu.data, filters["F469N"])
f469nc = mean_profile(hdu.data, filters["F469N shift"])
```

```python
plot(xpos*3600, f547m_rubin/f547m, lw=0.5, alpha=0.7, label="F547M Red/Full")
plot(xpos*3600, fq575n/f547m_rubin, lw=0.5, alpha=0.7, label="FQ575N / (Red half F547M)")
plot(xpos*3600, f469n/f547m, lw=0.5, alpha=0.7, label="F469N/F547M")
plot(xpos*3600, f469nc/f547m, lw=0.5, alpha=0.7, label="F469N(cont)/F547M")
legend()
ylim(0.8, 1.5)
grid()
gcf().set_size_inches(20,8)
```

```python
def double_up(a):
    n = len(a)
    if n%2 == 1:
        # pad to even length
        a = np.array(a.tolist() + [a[-1]])
        n += 1
    return a.reshape((n/2, 2)).mean(axis=1)

b = np.arange(13)
print b
print double_up(b)
```

```python
x = xpos*3600
good_mask = (f547m > 0.8*np.median(f547m)) & (np.abs(np.abs(x) - 11.75) > 0.45)
y1 = fq575n[good_mask]/np.median(f547m)
y2 = f547m[good_mask]/np.median(f547m)
xm = x[good_mask]
factor = 0.938
label1 = "FQ575N continuum"
label2 = "{:.3f} x F547M continuum".format(factor)
fill_between(x, (~good_mask).astype(int)*5.0, facecolor="k", alpha=0.2)
for i in range(6):
    plot(xm, y1 + 0.1*i, "b", label=label1)
    plot(xm, factor*y2 + 0.1*i, "r.", label=label2, alpha=0.4)
    xm = double_up(xm)
    y1 = double_up(y1)
    y2 = double_up(y2)
    label1, label2 = None, None
ylim(0.6, 1.8)
xlim(-25, 25)
legend()
title("Bally HST10 G430L slit: Broad versus narrow continuum comparison as a function of binning")
xlabel("Offset, arcsec")
ylabel("Continuum intensity + Offset")
gcf().set_size_inches(12,6)
gcf().savefig("bally-continuum-comparison-with-binning.pdf")
```

### Look for correlations between the ratios

```python
import scipy.stats
```

First do it with the filter shifted slightly to avoid the emission lines.

```python
F1, F2, F3 = "F469N shift", "FQ575N shift", "FQ575N double shift"
width = 0.22
x = xpos*3600
good_mask = (f547m > 0.8*np.median(f547m)) & (np.abs(np.abs(x) - 11.75) > 0.45) & (xpos < 0.0068786014) & (xpos > -0.0071414223)
#good_mask = good_mask & (np.abs(x) > 2.0) # exclude proplyd
rat1 = mean_profile(hdu.data, filters[F1])[good_mask]/f547m[good_mask]
x1 = np.median(rat1) - 0.5*width
x2 = np.median(rat1) + 0.5*width
rat2 = mean_profile(hdu.data, filters[F2])[good_mask]/f547m[good_mask]
y1 = np.median(rat2) - 0.5*width
y2 = np.median(rat2) + 0.5*width
rat3 = mean_profile(hdu.data, filters[F3])[good_mask]/mean_profile(hdu.data, filters[F2])[good_mask]
alpha = 0.3
ms = 4.0
label1 = "{} / F547M".format(F1.split()[0])
label2 = "{} / F547M".format(F2.split()[0])
var_rat1, var_rat2, var_rat3 = [], [], []
bins = np.arange(8)
for i in bins:
    print "{} x binning".format(2**i)
    var_rat1.append(np.var(rat1))
    var_rat2.append(np.var(rat2))
    var_rat3.append(np.var(rat3))
    print "{} = {:.4f} +/- {:.4f}, {} = {:.4f} +/- {:.4f}".format(label1, np.median(rat1), np.std(rat1), 
                                                                  label2, np.median(rat2), np.std(rat2))
    plot(rat1, rat2, "o", alpha=alpha, ms=ms, 
         label=r"${} \times$ binning".format(2**i))
    slope, intercept, r_value, p_value, std_err = scipy.stats.linregress(rat1, rat2)
    print "Linear fit: y = {:.3f} x + {:.3f}, r = {:.3f}, p = {:.3e}, R^2 = {:.3f}".format(
                slope, intercept, r_value, p_value, r_value**2)
    rat1 = double_up(rat1)
    rat2 = double_up(rat2)
    rat3 = double_up(rat3)
    ms *= 1.2
    alpha += 0.05
    print
plot(rat1, rat2, 'o', alpha=0.5)
xlim(x1, x2)
ylim(y1, y2)
legend()
grid()
title("Correlation in continuum bands")
xlabel(label1)
ylabel(label2)
gcf().set_size_inches(8,8)
```

So the upshot is that F469N/F547M is almost useless for predicting FQ575N(cont)/F547M.

The slope is about -0.1 but the R-squared value is only 3% (at all scales) or 20% (after binning) , which is the fraction of the variance in FQ575N(cont)/F547M that can be explained by variations in F469N/F547M.  The s.d. of F469N/F547M is much larger: 3% versus < 1% for FQ575N(cont)/F547M (after binning).  If we include the proplyd as well, then the slope becomes very slightly larger, but R-squared is still < 30%.

```python
loglog(2**bins, var_rat1, 'b-o', lw=4, ms=8, label="F469N / F547M")
loglog(2**bins, var_rat2, 'g-o', lw=4, ms=8, label="FQ575N / F547M")
loglog(2**bins, np.array(var_rat3), 'r-o', lw=4, ms=8, label="FQ575N' / FQ575N")
loglog(2**bins, var_rat2 - 0.5*np.array(var_rat3), 'g--o', lw=4, ms=8, label="Noise-subtracted")
loglog(2**bins, var_rat1 - 0.542*(20./18.)*np.array(var_rat3), 'b--o', lw=4, ms=8, label="Noise-subtracted")
# noise_var = np.media
# noise_var = np.median(err_nii_cont[good_mask]/profile_nii_cont[good_mask])**2 + np.median(err_f547m[good_mask]/profile_f547m[good_mask])**2
noise_var = 4.5e-4
loglog(2**bins, noise_var/2**bins, "c", lw=2, label=r"Noise $\propto\ N^{-1}$")
a = 0.03
loglog(2**bins, noise_var*(a + (1 - a)/2**bins), "c--", lw=2, label=r"Noise $+\ C$")
xlabel("bin size, $N$, pixels")
ylabel(r"Variance, $\sigma^2$")
legend(loc="upper right")
ylim(1e-5, None)
grid()
gcf().set_size_inches(7, 7)
```

This shows how the variance in the ratios falls with binning.  The dark blue is the 4690 narrow–broad continuum ratio, which maintains its high variance as the binning increases, indicating that it is real structure on largish scales.  

The red line is for two adjacent continuum sectons.  The variance between these should be dominated by noise.  Indeed it is the steepest observed curve, and it is close to $N^{-1}$ in the middle section.  It does start to flatten off at large binning – possibly we are hitting some sort of pattern noise.

The green line is the variance of the 5755 narrow–broad continuum ratio, which is what we are most interested in.  This is steeper than the line for the blue continuum, but shallower than the "noise" line.  The two narrow bands FQ575N and FQ575N' are of the same width (20 pix), while F547M is much larger (238 pix).  The noise contribution to the green line should therefore be (1/20 + 1/238) / (1/20 + 1/20) = 0.542 x the red line.  But if we use that, then we oversubtract the noise at small scales, so I use 0.5 instead.  This gives the green dashed line, which is remarkably flat, like the blue line.  This implies that the /true/ variation in the 575N/547M continuum ratio is all at largeish scales, with variance 2e-5 => s.d. of 0.45%. 

For comparison, the blue dashed line is a similar noise correction to the 469N/547M ratio, which hardly changes things at all.  So this also has a flat profile, with variance of 8e-4 => s.d. of 2.8%.  This is 6.3 times higher than for the green line.  The wavelength difference between the narrow filter and the center of the broad filter is 3 times higher for 469–547 than for 575–547.  There is no reason to expect these factors to be equal though. 

The fact that both the green and blue lines curve down at the highest binning may indicate that we are hitting the dominant scale of the variations and starting to integrate things away.  This would be around 100 pixels, or 5 arcsec. 

Now repeat, but using the true filter wavelength range, which probably includes too much of the emission lines.

```python
F1, F2= "F469N", "FQ575N shift"
width = 0.22
x = xpos*3600
good_mask = (f547m > 0.8*np.median(f547m)) & (np.abs(np.abs(x) - 11.75) > 0.45) & (xpos < 0.0068786014) & (xpos > -0.0071414223)
good_mask = good_mask & (np.abs(x) > 2.0) # exclude proplyd
rat1 = mean_profile(hdu.data, filters[F1])[good_mask]/f547m[good_mask]
x1 = np.median(rat1) - 0.5*width
x2 = np.median(rat1) + 0.5*width
rat2 = mean_profile(hdu.data, filters[F2])[good_mask]/f547m[good_mask]
y1 = np.median(rat2) - 0.5*width
y2 = np.median(rat2) + 0.5*width
alpha = 0.4
ms = 4.0
label1 = "{} / F547M".format(F1.split()[0])
label2 = "{} / F547M".format(F2.split()[0])
for i in range(6):
    print "{} x binning".format(2**i)
    print "{} = {:.4f} +/- {:.4f}, {} = {:.4f} +/- {:.4f}".format(label1, np.median(rat1), np.std(rat1), 
                                                                  label2, np.median(rat2), np.std(rat2))
    plot(rat1, rat2, "o", alpha=alpha, ms=ms, 
         label=r"${} \times$ binning".format(2**i))
    slope, intercept, r_value, p_value, std_err = scipy.stats.linregress(rat1, rat2)
    print "Linear fit: y = {:.3f} x + {:.3f}, r = {:.3f}, p = {:.3e}".format(slope, intercept, r_value, p_value)
    rat1 = double_up(rat1)
    rat2 = double_up(rat2)
    ms *= 1.2
    alpha += 0.1
    print
plot(rat1, rat2, 'o', alpha=0.5)
xlim(x1, x2)
ylim(y1, y2)
legend()
grid()
title("Correlation in continuum bands")
xlabel(label1)
ylabel(label2)
gcf().set_size_inches(8,8)
```

This is even flatter, and the correlation is even smaller.

So there is a systematic spatial variation in FQ575N(cont)/F547M of about 0.7% RMS that survives the binning: 0.938 +/- 0.007.  When we used a broader band for the 575 continuum we got reduced noise, but a systematically larger value, 0.945 +/- 0.005, because were extending the band to the blue.

There is also a difference of 0.01 between the Bally and Rubin values.  We don't know if this is also spatial variation, or if it represents systematic errors in the flux calibration of the two gratings (G430L and G750M).

All we can say is that this gives a fundamental limit of about 1% to the accuracy of the determination of the continuum at the 5755 line, even if there is no error in the F547M measurement.
