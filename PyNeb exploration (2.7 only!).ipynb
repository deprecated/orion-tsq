```python
cliii = pyneb.Atom('Cl', 3)
```

```python
def oii_den_ratio(den):
    x = np.log10(den)
    return 0.65*(np.tanh(1.2*(x - 3.45)) + 1.2)
```

```python
def Ddiag_plot(atom, wavA, wavB, temrange, denrange):
    for tem in temrange:
        A = atom.getEmissivity(tem, denrange, wave=wavA)
        B = atom.getEmissivity(tem, denrange, wave=wavB)
        plot(denrange, A/B, label="T = {:.0f} K".format(tem))
    legend(loc="upper right")
    xlabel("Density, pcc")
    ylabel("{} / {}".format(wavA, wavB))
    
def cliii_den_ratio(den):
    x = np.log10(den)
    return 0.55*(np.tanh(1.2*(3.55 - x)) + 1.62)

denrange = np.logspace(2.0, 5.0, 200)
Ddiag_plot(cliii, 5517, 5537, [6e3, 8e3, 1e4, 1.2e4], 
           denrange)
plot(denrange, oii_den_ratio(denrange))
plot(denrange, cliii_den_ratio(denrange), '.')
grid()
ylim(0.0, None)
xscale('log')
gcf().set_size_inches(8, 8)
```
