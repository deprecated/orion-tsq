```python
filters = {}
filters["FQ575N"] = {"wav0": 5758.0, "dwav": 20.6}
filters["F547M"] = {"wav0": 5547.0, "dwav": 650.0}
filters["5755"] = {"wav0": 5755.08 + 1.1, "dwav": 5.6}
filters
```

```python
for name, params in filters.items():
    params["i1"] = len(wavs[wavs < params["wav0"] - 0.5*params["dwav"]])
    params["i2"] = len(wavs[wavs <= params["wav0"] + 0.5*params["dwav"]]) + 1
filters["F547M"]["i1"] = 89
filters
```

```python
zero_mask = hdu.data < 0.0
hdu.data[zero_mask] = 0.0
f1 = filters["F547M"]
profile_f547m = hdu.data[:, f1["i1"]:f1["i2"]].mean(axis=1)
f2 = filters["5755"]
profile_nii_raw = hdu.data[:, f2["i1"]:f2["i2"]].mean(axis=1)
f3 = filters["FQ575N"]
profile_nii_cont = 0.5*(hdu.data[:, f3["i1"]:f2["i1"]].mean(axis=1) + hdu.data[:, f2["i2"]:f3["i2"]].mean(axis=1))
profile_nii = profile_nii_raw - 0.93*profile_f547m
profile_nii_old = profile_nii_raw - profile_nii_cont 
plot(xpos*3600, profile_nii, label="line")
plot(xpos*3600, profile_nii_cont, label="cont")
plot(xpos*3600, 0.93*profile_f547m, label="0.93 x wide cont")
plot(xpos*3600, profile_nii_old, "k", lw=0.2, label=None)
xlim(-28.0, 28.0)
ylim(0.0, 6.e-15)
legend()
gcf().set_size_inches(15, 6)
```

Previously I had been using the immediately adjacent wavelengths to do the continuum subtraction.  But it turns out (see below) that it is much better to use a very broad wavelength range to estimate the continuum.  The reduction in noise so-obtained vastly outweighs the uncertainty in the color correction.  In a later section, I determine the color correction to be 0.93.

#### Equivalent width of [N II] 5755

```python
ew_nii = (f2["i2"] - f2["i1"])*dwav*profile_nii/profile_nii_cont
ew_nii_estimate = (f2["i2"] - f2["i1"])*dwav*profile_nii/(0.93*profile_f547m)
plot(xpos*3600, ew_nii, label="True EW(5755)")
plot(xpos*3600, ew_nii_estimate, label="Estimated EW(5755)")
xlim(None, None)
ylim(0.0, 40.)
legend()
gcf().set_size_inches(20, 6)
```

So this is actually much better than I was getting in more outer parts of the nebula (EW ~ 3 Angstrom).  Here it is 5 to 15 Angstrom.  Since the FQ575N filter width is 20.6 Angstrom, this means that 20% to 42% of the signal in the filter is actual line.  So the continuum correction is not so overwhelming as I had previously thought.

#### Profile of the [N II] 5755 by the two methods

This compares the "true" spatial profile of 5755 (best estimate from the spectrum) with the "broad-band" estimation of the same, which is a simulation of what we can obtain from the FQ575N and F547M WFC3 filters.

```python
profile_nii_true = (f2["i2"] - f2["i1"])*dwav*profile_nii
e_nii_true = hdulist["ERR"].data[:, f2["i1"]:f2["i2"]].sum(axis=1)*dwav/np.sqrt(f2["i2"] - f2["i1"]) 
profile_nii_estimate = hdu.data[:, f3["i1"]:f3["i2"]].sum(axis=1)*dwav - 0.93*profile_f547m*(f3["i2"] - f3["i1"])*dwav
plot(xpos*3600, profile_nii_true, '-o', label="True 5755", alpha=0.6)
plot(xpos*3600, profile_nii_estimate, '.-', label="Estimated 5755", alpha=0.6)
plot(xpos*3600, e_nii_true, label="Error 5755")
xlim(-25.0, 25.0)
ylim(0.0, 4e-14)
legend()
xlabel("Slit offset, arcsec")
ylabel("Line flux, cgs")
gcf().tight_layout()
gcf().set_size_inches(30, 6)
gcf().savefig("stis-5755-profile.pdf")
```

The coefficient of determination $R^2$ is the fraction of the variance in $y$ that is explained by the variance in $x$.

```python
print "For all data, R-squared = ", r_value**2
```

That is *now great*. Try again, but selecting the best parts of the slit.

```python
slope, intercept, r_value, p_value, std_err = scipy.stats.linregress(x[mm & m], y[mm & m])
print slope, intercept, r_value, p_value, std_err
```

```python
print "For selected data, R-squared = ", r_value**2
```

That is even better: 60% of the variance in the true ratio can be predicted by the estimated ratio.   Just to be sure, we will also do it the other way around.


```python
slope, intercept, r_value, p_value, std_err = scipy.stats.linregress(y[mm & m], x[mm & m])
print slope, intercept, r_value, p_value, std_err
print "For selected data, R-squared = ", r_value**2
```

### Try spatial binning to see if the correlation improves

```python
def double_up(a):
    n = len(a)
    if n%2 == 1:
        # pad to even length
        a = np.array(a.tolist() + [a[-1]])
        n += 1
    return a.reshape((n/2, 2)).mean(axis=1)

b = np.arange(13)
print b
print double_up(b)
```

The variance of the individual continua are almost constant, since they are dominated by real physical structure.  This is especially true for F547M where the noise is extremely low (since we are averaging over a large wavelength range).  For the FQ575N continuum, we are averaging over narrow strips either side of the 5755 line, so the noise is higher, and increases the variance for small $N$.  

The variance of the ratio, on the other hand, falls very sharply as the binning is increased.  

Note that there are no free parameters in that "noise" line.  The noise magnitude for 1-pixel binning is calculated from the "ERR" HDU of the STIS FITS file, assuming no pixel-pixel correlations in the dispersion dimension.  So, it is heartening that it agrees so well with the data, although it is odd that the variance falls slower that $1/N$ at the smallest scales.  So, at all scales the noise contributes at least 75% of the total variance in FQ575N(Cont)/F547M.  If there is another contribution to the variance, then it also behaves as though it were spatially uncorrelated. 

The dashed line shows the noise-corrected behavior, but this is very dubious because it depends critically on the noise being well-behaved.  In the next section, I show that the noise is not always well-behaved.  The upshot is that it is likely that *all* variations in the continuum ratio are due to noise and *there is no evidence for systematic spatial variations in the continuum color*.  

So the best thing to do is probably to apply the constant color term of 0.93 to the [N II] 5755 line.

So this is a bit of a cautionary tale.  The nominal errors from the FITS file give a very good estimate of the variance per pixel.  But the fall-off with binning is not as $N^{-1}$ but more like $N^{-0.75}$.  Perhaps this is due to correlations in the noise?  It is strange that this ratio is less well behaved than the previous one, given that I would expect this to be pure noise.
