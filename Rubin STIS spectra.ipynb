Reject cosmic rays via simple threshold of the difference of the two exposures

```python
pattern = "/Users/will/Work/HST-STIS/Rubin7514/o4fo{}_x2d.fits"
hdulist1 = fits.open(pattern.format("020h0"))
hdulist2 = fits.open(pattern.format("020i0"))
hdu1 = hdulist1["SCI"]
ehdu1 = hdulist1["ERR"]
hdu2 = hdulist2["SCI"]
ehdu2 = hdulist2["ERR"]
merge = 0.5*(hdu1.data + hdu2.data)
emerge = 0.5*np.sqrt(ehdu1.data**2 + ehdu2.data**2)
THRESHOLD = 1e-14
m = hdu1.data - hdu2.data > THRESHOLD
merge[m] = hdu2.data[m]
emerge[m] = ehdu2.data[m]
m = hdu2.data - hdu1.data > THRESHOLD
merge[m] = hdu1.data[m]
emerge[m] = ehdu1.data[m]
hdu1.data = merge
ehdu1.data = emerge
hdulist1.writeto("rubin-stis-slit1-5734-merge.fits", clobber=True)
```

```python
hdu = hdulist["SCI"]
nx, wav0, i0, dwav = [hdu.header[k] for k in "NAXIS1", "CRVAL1", "CRPIX1", "CD1_1"]
wavs = wav0 + (np.arange(nx) - (i0 - 1))*dwav 
ny, x0, j0, dx = [hdu.header[k] for k in "NAXIS2", "CRVAL2", "CRPIX2", "CD2_2"]
xpos = x0 + (np.arange(ny) - (j0 - 1))*dx 
```

```python
import aplpy
```

```python
f = aplpy.FITSFigure(hdu)
f.show_grayscale(vmin=1e-16, stretch="log")
f.show_colorbar()
```

```python
f = aplpy.FITSFigure(hdulist["ERR"])
f.show_grayscale(vmin=1e-16, vmax=1.327e-13, stretch="log")
f.show_colorbar()
```

```python
filters = {}
filters["FQ575N"] = {"wav0": 5758.0, "dwav": 20.6}
filters["F547M"] = {"wav0": 5547.0, "dwav": 650.0}
filters["5755"] = {"wav0": 5755.08 + 1.25, "dwav": 5.5}
filters["6584"] = {"wav0": 6583.45 + 1.25, "dwav": 5.5}
filters
```

```python
for name, params in filters.items():
    params["i1"] = len(wavs[wavs < params["wav0"] - 0.5*params["dwav"]])
    params["i2"] = len(wavs[wavs <= params["wav0"] + 0.5*params["dwav"]]) + 1
filters
```

```python
f1 = filters["F547M"]
profile_f547m = hdu.data[:, f1["i1"]:f1["i2"]].mean(axis=1)
f2 = filters["5755"]
profile_nii = hdu.data[:, f2["i1"]:f2["i2"]].mean(axis=1)
f3 = filters["FQ575N"]
profile_nii_cont = 0.5*(hdu.data[:, f3["i1"]:f2["i1"]].mean(axis=1) + hdu.data[:, f2["i2"]:f3["i2"]].mean(axis=1))
profile_nii -= profile_nii_cont 
plot(xpos*3600, profile_nii, label="line")
plot(xpos*3600, profile_nii_cont, label="cont")
plot(xpos*3600, profile_f547m, label="wide cont")
xlim(-28.0, 28.0)
ylim(0.0, 6.e-15)
legend()
gcf().set_size_inches(15, 6)
```

```python
ew_nii = (f2["i2"] - f2["i1"])*dwav*profile_nii/profile_nii_cont
ew_nii_estimate = (f2["i2"] - f2["i1"])*dwav*profile_nii/profile_f547m
plot(xpos*3600, ew_nii, label="True EW(5755)")
plot(xpos*3600, ew_nii_estimate, label="Estimated EW(5755)")
xlim(None, None)
ylim(0.0, 40.)
legend()
gcf().set_size_inches(20, 6)
```

```python
profile_nii_true = (f2["i2"] - f2["i1"])*dwav*profile_nii
e_nii_true = hdulist["ERR"].data[:, f2["i1"]:f2["i2"]].sum(axis=1)*dwav/np.sqrt(f2["i2"] - f2["i1"]) 
profile_nii_estimate = hdu.data[:, f3["i1"]:f3["i2"]].sum(axis=1)*dwav - profile_f547m*(f3["i2"] - f3["i1"])*dwav
plot(xpos*3600, profile_nii_true, '-o', label="True 5755", alpha=0.6)
plot(xpos*3600, profile_nii_estimate, '.-', label="Estimated 5755", alpha=0.6)
plot(xpos*3600, e_nii_true, label="Error 5755")
xlim(-25.0, 25.0)
ylim(0.0, 4e-14)
legend()
xlabel("Slit offset, arcsec")
ylabel("Line flux, cgs")
gcf().tight_layout()
gcf().set_size_inches(30, 6)
gcf().savefig("stis-5755-profile.pdf")
```
