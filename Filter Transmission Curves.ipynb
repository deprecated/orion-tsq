```python
import os
def get_filter(fname, UVIS=1, return_wavelength=False):
    datafile = "{:s}.UVIS{:01d}.tab".format(fname.lower(), UVIS)
    fullpath = os.path.join("WFC3-filters", "SystemThroughput", datafile) 
    data = genfromtxt(fullpath,
                      names=("row", "wavelength", "throughput"))
    if return_wavelength:
        return data["wavelength"], data["throughput"]
    else:
        return data["throughput"]
```

```python
wavs, f673n = get_filter("F673N", UVIS=1, return_wavelength=True)
f673n2 = get_filter("F673N", UVIS=2)
fq674n = get_filter("FQ674N", UVIS=1)
fq674n2 = get_filter("FQ674N", UVIS=2)
fq672n = get_filter("FQ672N", UVIS=1)
fq672n2 = get_filter("FQ672N", UVIS=2)
```

```python
plot(wavs, f673n)
plot(wavs, f673n2)
xlim(6680.0, 6880.0)
xlabel("wavelength")
title("Comparison of UVIS1 and UVIS2 transmission curves")
```

```python
plot(wavs, f673n/f673n2)
plot(wavs, fq672n/fq672n2)
plot(wavs, fq674n/fq674n2)
xlim(6500.0, 7000.0)
ylim(0.9, 1.1)
```
